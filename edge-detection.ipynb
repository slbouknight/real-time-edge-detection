# Edge Detection with Live Webcam

This notebook experiments with various methods of edge detection including Sobel, Laplacian, and Canny. Each method can be visualized in real time via a webcam.

## Imports

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

## Functions for Different Edge Detection Methods

```python
def sobel_y_detection(color_image):
    '''Takes in a color image and performs Sobel vertical edge detection'''
    
    # Converting provided image to grayscale
    gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)
    sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=7)
    
    return sobel_y
```

```python
def sobel_x_detection(color_image):
    '''Takes in a color image and performs Sobel horizontal edge detection'''
    
    # Converting provided image to grayscale
    gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)
    sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=7)
    
    return sobel_x
```

```python
def laplacian_detection(color_image):
    '''Takes in a color image and performs Laplacian edge detection'''
    
    # Converting provided image to grayscale
    gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)
    laplacian = cv2.Laplacian(gray_image, cv2.CV_64F)
    
    return laplacian
```

```python
def canny_detection(color_image):
    '''Takes in a color image and performs Canny edge detection'''
    
    threshold_1 = 30
    threshold_2 = 80
    
    # Converting provided image to grayscale
    gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)
    canny = cv2.Canny(gray_image, threshold_1, threshold_2)
    
    return canny
```

## Initialize Webcam and Apply Edge Detection

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read() # Cap.read() returns a ret bool to indicate success.
    
    # Display webcam footage with each detection displayed
    cv2.imshow('Sobel Vertical Edge Detection', sobel_y_detection(frame))
    cv2.imshow('Sobel Horizontal Edge Detection', sobel_x_detection(frame))
    cv2.imshow('Laplacian Edge Detection', laplacian_detection(frame))
    cv2.imshow('Canny Edge Detection', canny_detection(frame))
    cv2.imshow('Webcam Video', frame)
    
    # Press enter key to exit loop
    if cv2.waitKey(1) == 13:
        break
    
# Stop capturing footage and close windows    
cap.release()
cv2.destroyAllWindows()      
```
